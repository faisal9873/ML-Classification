```python
df3.head(6)
```

```python
shape = df3.shape
print('\nnum samples :', shape[0])
print('\nnum attributes :', shape[1])
```

```python
df4 = df3.drop(['PassengerId', 'Name', 'Cabin', 'Ticket'], axis=1)
df4.head(6)
```

```python
df5 = pd.get_dummies(df4)
print(df5.shape)
df5.head(6)
```

```python
df5.dropna(inplace=True) #dropping rows with missing data
print(df5.shape)
print('number of rows deleted: ', df4.shape[0]-df5.shape[0])
```

```python
# separate features and target
y = df5['Survived']
X = df5.drop(['Survived'], axis=1)

# split the data into training and test set
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
```

#Fitting the Logistic Regression Model

```python
#Logisitic Regression
logreg = LogisticRegression(max_iter=1000,random_state=42)

# train the model
logreg.fit(X_train, y_train)

# make predictions on test set
y_pred = logreg.predict(X_test)

# measure accuracy
score = accuracy_score(y_test, y_pred)
print("Accuracy:", score)
```

# Confusion Matrix



```python
confusion_matrix(y_test, y_pred)
```
